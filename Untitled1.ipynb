```python
import pandas as pd
import string
import nltk


nltk.download('punkt_tab')

# Download NLTK resources (Run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Load dataset
df = pd.read_csv("Twitter Sentiments.csv")

# Print column names
print("Columns in dataset:")
print(df.columns.tolist())

# Automatically find the text column
possible_columns = ['text', 'tweet', 'tweets', 'message', 'content', 'Text', 'Tweet']

text_column = None
for col in possible_columns:
    if col in df.columns:
        text_column = col
        break

# If none found, use the second column (usually tweets)
if text_column is None:
    text_column = df.columns[1]

print("Using text column:", text_column)

# Stopwords
stop_words = set(stopwords.words('english'))

# Lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to remove only 'ing'
def remove_ing(word):
    if word.endswith("ing") and len(word) > 4:
        return word[:-3]
    return word

# Preprocessing function
def preprocess(text):

    if pd.isna(text):
        return ""

    # 1. Lowercase
    text = str(text).lower()

    # 2. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 3. Tokenization
    tokens = word_tokenize(text)

    # 4. Stopword removal
    tokens = [word for word in tokens if word not in stop_words]

    # 5. Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    # 6. Remove only "ing"
    tokens = [remove_ing(word) for word in tokens]

    return " ".join(tokens)

# Apply preprocessing
df["Processed_Text"] = df[text_column].apply(preprocess)

# Display first five rows
print(df[[text_column, "Processed_Text"]].head())

# Save output
df.to_csv("Twitter_Sentiments_Preprocessed.csv", index=False)

print("\nPreprocessing completed successfully!")
print("Saved as Twitter_Sentiments_Preprocessed.csv")
```
